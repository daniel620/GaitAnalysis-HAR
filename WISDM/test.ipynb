```python
def myDrawLine(x,title=''):
    if title:
        plt.suptitle(title)
    # x = xTrain[0]
    d = dict(zip(np.arange(6), [1,3,5,2,4,6])) 
    for i in range(6):
        plt.subplot(3,2,d[i])
        plt.plot(x[:,i])

    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
for j in range(5):
    i = j * 30 + 2
    print(i)
    myDrawLine(xTrain[i],title=str(yTrain[i]))
```
